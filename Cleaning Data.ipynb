# Outerlier Detecting

```python
columns_to_check = ['price', 'minimum_nights']

# Function to detect outliers using IQR
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detecting outliers
outliers_price = detect_outliers_iqr(df, 'price')
outliers_min_nights = detect_outliers_iqr(df, 'minimum_nights')

print("Outliers in Price:", outliers_price.shape[0])
print("Outliers in Minimum Nights:", outliers_min_nights.shape[0])
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Keeping only non-outlier data
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_filtered

# Apply to both columns
df_cleaned = remove_outliers_iqr(df, 'price')
df_cleaned = remove_outliers_iqr(df_cleaned, 'minimum_nights')

print("Shape after outlier removal:", df_cleaned.shape)
```
